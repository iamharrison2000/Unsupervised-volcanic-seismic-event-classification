```python
from tensorflow.python.summary.summary_iterator import summary_iterator
import matplotlib.pyplot as plt
import numpy as np
```

```python
summary = summary_iterator("./runs/Nov09_17-35-17_SDSS-F4GMG72RN2/events.out.tfevents.1668044117.SDSS-F4GMG72RN2.4396.0")
```

```python
train_loss = []
val_loss =[]
train_acc = []
val_acc = []
for items in summary_iterator("./runs//Nov09_20-47-48_SDSS-F4GMG72RN2/events.out.tfevents.1668055668.SDSS-F4GMG72RN2.12231.0"):
    for v in items.summary.value:
        if v.tag == 'train_loss':
            train_loss.append(v.simple_value)
        if v.tag == 'val_loss':
            val_loss.append(v.simple_value)
        if v.tag == 'train_acc':
            train_acc.append(v.simple_value)
        if v.tag == 'val_acc':
            val_acc.append(v.simple_value)
```

```python
epoch = np.linspace(1,len(train_loss),len(train_loss))

fig, ax = plt.subplots(figsize=(5,3),dpi=200)
# ax.set_yscale('log')
ax.plot(epoch, train_loss, label='train_loss')
ax.plot(epoch, val_loss, label='val_loss')
# ax.plot(epoch, recon_loss, label='recon_loss')
ax.legend(fontsize=5)
ax.tick_params(axis='both', which='major', labelsize=8)
ax.tick_params(axis='both', which='minor', labelsize=8)
ax.set_title('loss')
ax.set_xlabel('epochs')
plt.show()

fig, ax = plt.subplots(figsize=(5,3),dpi=200)
# ax.set_yscale('log')
ax.plot(epoch, train_acc, label='train_acc')
ax.plot(epoch, val_acc, label='val_acc')
ax.legend(fontsize=5)
ax.tick_params(axis='both', which='major', labelsize=8)
ax.tick_params(axis='both', which='minor', labelsize=8)
ax.set_title('acc')
ax.set_xlabel('epochs')
plt.show()
```
